```python
df_country = pd.read_excel('Country-Code.xlsx')
```

```python
[features for features in df.columns if df[features].isnull().sum()>0]
```

```python
df=pd.merge(df, df_country, how='left', on = 'Country Code')
```

```python
df.head()
```

```python
df.info()
```

```python
country_names = df['Country'].value_counts().index
country_names
country_values = df['Country'].value_counts().values
country_values
```

```python
plt.pie(country_values[:3], labels=country_names[:3], autopct="%0.2f%%")
```

**India has most number of orders 94.39% and then USA and then UK**

```python
rating_df = df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
rating_df
```

```python
plt.figure(figsize=(14, 8))
sns.barplot(x='Aggregate rating', y='Rating count', data=rating_df)
```

**Observation**
- Max items are not rated
- Max no. of items are rated as 3.1-3.5
